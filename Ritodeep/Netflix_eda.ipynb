```python
df1.reset_index(inplace=True)
df1.head()
```

Resetting The Index Numbers

```python
rec=[]
for i in range(len(df1['user rating score'])):
    if df1['user rating score'][i]>=90.0:
        rec.append('A must watch')
    if df1['user rating score'][i]>=80.0 and df1['user rating score'][i]<90.0:
        rec.append("Highly reccomended")
    if df1['user rating score'][i]>=70.0 and df1['user rating score'][i]<80.0:
        rec.append('Recommended')
    if df1['user rating score'][i]>=50.0 and df1['user rating score'][i]<70:
        rec.append('A one time watch')
    if df1['user rating score'][i]<50.0:
        rec.append('Can be skipped')
df1['Recommendation']=rec
df1
```

Adding a new column recommendation which recommends movies based on user rating

```python
df3=df1.sort_values(by=['release year'],ascending=False)
df3
```

Arranging the shows based on their release year (newest to oldest)

```python
df4=df1.sort_values(by=['user rating score'],ascending=False)
df4
```

```python

plt.figure(figsize=(50,20))
plt.title("Graph of movie released in an year",fontsize=50)
x=df1['release year']
y=df1['user rating score']
plt.plot(x,y,linewidth=3)
plt.xlabel("Year",fontsize=50)
plt.ylabel("User rating Score",fontsize=50)
plt.show()
```

```python
plt.figure(figsize=(30,20))
plt.title("Graph of movie released in an year",fontsize=50)
x=df1['release year']
y=df1['user rating score']

plt.xlabel("Year",fontsize=50)
plt.ylabel("'user rating score'",fontsize=50)
plt.bar(x,y,width = 1, color = ['red', 'green'])
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.title("Rating vs Shows released in a particular year", fontsize = 20)

x=df1['release year']
y=df1['user rating score']

plt.xlabel('Shows released in a year',fontsize=15)
plt.ylabel('User rating Score',fontsize=15)
plt.scatter(x, y,alpha = 0.5)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title("Graph of movie released in an year",fontsize=20)
rel=df1['release year'].value_counts()
plt.xlabel("Year",fontsize=15)
plt.ylabel("Number of movies release",fontsize=15)
rel.plot.line(linewidth=3)
```
