```python
rating_freq = df['rating'].value_counts()
plt.figure(figsize=(40,20))
rating_freq.plot.bar(color='Red',alpha=0.5)
plt.xlabel("Ratings")
plt.ylabel("Freq of ratings")
plt.title("RATING")
```

```python
from matplotlib import style
style.use('ggplot')
rating_freq = df['rating'].value_counts()
plt.figure(figsize=(20,10))
rating_freq.plot.line(color='Red',linewidth=4)
plt.xlabel("Ratings")
plt.ylabel("Freq of ratings")
plt.title("RATING")
```

```python
rating_freq = df['rating'].value_counts()
plt.figure(figsize=(20,10))
rating_freq.plot.pie()
#plt.xlabel("Ratings")
#plt.ylabel("Freq of ratings")
plt.title("RATING")
```

```python
year_freq = df['release year'].value_counts().sort_values(ascending= True)
plt.figure(figsize=(20,10))
year_freq.plot.bar(color='green')
plt.xlabel("Release Years")
plt.ylabel("Movies Released")
plt.title("Movies Released Vs Release Years: Bar Graph")
plt.show()
```

```python
year_freq = df['release year'].value_counts().sort_values(ascending= True)
plt.figure(figsize=(20,10))
year_freq.plot.pie()
#plt.xlabel("Release Years")
#plt.ylabel("Movies Released")
plt.title("Movies Released Vs Release Years: Pie Chart")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
year_freq = df['release year'].value_counts().sort_values(ascending= True)
plt.figure(figsize=(20,10))
year_freq.plot.line(color='purple' ,linewidth=4)
plt.xlabel("Release Years")
plt.ylabel("Movies Released")
plt.title("Movies Released Vs Release Years: Graph")
plt.show()
```
